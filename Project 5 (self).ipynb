```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
len(stem_cats)+len(lib_arts_cats)+len(other_cats)
```

```python
fig=plt.figure(figsize=(18,12))

for i in range(0,16,3):
    ax=fig.add_subplot(6,3,i+1)
    ind=int(i/3)
    ax.plot(women_degrees['Year'],women_degrees[stem_cats[ind]], label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[ind]], label='Men')
    ax.set_title(stem_cats[ind])
    ax.set_ylim(0,100)
    ax.set_xlim(1970,2010)
    ax.axhline(50, alpha= 0.5)
    ax.tick_params(bottom=False, labelbottom=False)
    if ind==5:
        ax.tick_params(labelbottom=True)
    if i==0:
        ax.text(2005,10, 'Women')
        ax.text(2006, 90,'Men')
    elif ind==5 :
        ax.text(2005,85, 'Women')
        ax.text(2006, 17, 'Men')
    
        
        
for b in range(1,16,3):
    ax=fig.add_subplot(6,3,b+1)
    inde=int(b/3)
    ax.plot(women_degrees['Year'],women_degrees[lib_arts_cats[inde]], label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[inde]], label='Men')
    ax.set_title(lib_arts_cats[inde])
    ax.set_ylim(0,100)
    ax.set_xlim(1970,2010)
    ax.tick_params(bottom=False, labelbottom=False)
    if inde==4:
        ax.tick_params(labelbottom=True)
    if b==0:
        ax.text(2005,10, 'Women')
        ax.text(2006, 90,'Men')
    elif inde==5:
        ax.text(2005,85, 'Women')
        ax.text(2006, 17, 'Men')
    
        

for a in range(2,18,3):
    ax=fig.add_subplot(6,3,a+1)
    index=int(a/3)
    ax.plot(women_degrees['Year'],women_degrees[other_cats[index]], label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[index]], label='Men')
    ax.set_title(other_cats[index])
    ax.set_ylim(0,100)
    ax.set_xlim(1970,2010)
    ax.tick_params(bottom=False, labelbottom=False)
    if inde==5:
        ax.tick_params(labelbottom=True)
    
plt.show()
```
